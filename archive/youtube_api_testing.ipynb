```python
# Dependencies
from config import yt_api_key as api_key
from pprint import pprint
import requests
import pandas as pd
```

```python
# Youtube API url and parameters
base_url = "https://www.googleapis.com/youtube/v3/search/"
query = "education"
params = {
    "part":"snippet",
    "q": query,
    "type": "video",
    "maxResults": 100,
    "order": "viewCount",
    "key": api_key
}
```

```python
# Get top 50 education channels
edu_chan = requests.get(base_url, params=params).json()
```

```python
pprint(edu_chan)
```

```python
# Store the channel IDs and titles
items = edu_chan["items"]
edu_chan_title_list = []
edu_chan_id_list = []

for i in range(len(edu_chan["items"])):
    edu_chan_id_list.append(items[i]["snippet"]["channelId"])
    edu_chan_title_list.append(items[i]["snippet"]["channelTitle"])
```

```python
# Create dataframe and drop duplicates
edu_df = pd.DataFrame()
edu_df["Title"] = edu_chan_title_list
edu_df["ID"] = edu_chan_id_list
edu_df.drop_duplicates()
```
